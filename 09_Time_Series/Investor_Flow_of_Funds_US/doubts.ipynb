```python
import pandas as pd

url = 'https://raw.githubusercontent.com/datasets/investor-flow-of-funds-us/master/data/weekly.csv'

df_funds=pd.read_csv(url, parse_dates=['Date'], index_col=0)
df_funds
```

```python
df_funds.resample('M').sum()
```

```python
dfres = df_funds.resample('Y').sum()
```

```python
dfres.style.background_gradient()
```
